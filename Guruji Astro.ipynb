```python
# Identify numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Summary statistics for key numeric columns
summary_stats = df[numeric_columns].describe()
print("Summary statistics for numeric columns:")
print(summary_stats)
```

```python


# Visualize the distribution of call charges
plt.figure(figsize=(10, 6))
sns.histplot(df['amount'], kde=True, bins=30)
plt.title('Distribution of Call Charges')
plt.xlabel('Call Charges (amount)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the average TalkTime for different call activities
if 'Activity' in df.columns and 'timeDuration' in df.columns:
    avg_talktime_per_activity = df.groupby('Activity')['timeDuration'].mean()
    print("Average TalkTime for different call activities:")
    print(avg_talktime_per_activity)
else:
    print("Columns 'Activity' or 'timeDuration' not found in the dataset.")
```

```python
# Determine the most common source of calls
if 'source' in df.columns:
    most_common_source = df['source'].value_counts().idxmax()
    print(f"The most common source of calls is: {most_common_source}")
else:
    print("Column 'source' not found in the dataset.")
```

```python
# Calculate total earnings and spending for both users and masters
if 'userOnCallDuration' in df.columns and 'amount' in df.columns:
    total_earnings_users = df[df['userOnCallDuration'] > 0]['amount'].sum()
    total_earnings_masters = df[df['astrologerOnCallDuration'] > 0]['astrologersEarnings'].sum()
    print(f"Total earnings (Users): {total_earnings_users}")
    print(f"Total earnings (Masters): {total_earnings_masters}")
else:
    print("Columns for earnings calculation not found in the dataset.")
```

```python
# Analyze the relationship between TalkTime and Charge
if 'timeDuration' in df.columns and 'amount' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='timeDuration', y='amount', data=df)
    plt.title('Relationship between TalkTime and Charge')
    plt.xlabel('TalkTime (minutes)')
    plt.ylabel('Charge (amount)')
    plt.show()
else:
    print("Columns for relationship analysis not found in the dataset.")
```

```python
# Calculate the average time it takes for calls to be connected
if 'ConnectTime' in df.columns and 'DialTime' in df.columns:
    df['ConnectDelay'] = df['ConnectTime'] - df['DialTime']
    avg_connect_delay = df['ConnectDelay'].mean()
    print(f"Average time for calls to be connected: {avg_connect_delay:.2f} seconds")
else:
    print("Columns 'ConnectTime' or 'DialTime' not found in the dataset.")
```

```python
# Identify the most common reason for call disconnection
if 'UnconnectTime' in df.columns and 'ConnectTime' in df.columns:
    df['UnconnectDelay'] = df['UnconnectTime'] - df['ConnectTime']
    most_common_disconnect_reason = df['UnconnectDelay'].value_counts().idxmax()
    print(f"The most common reason for call disconnection (delay): {most_common_disconnect_reason}")
else:
    print("Columns 'UnconnectTime' or 'ConnectTime' not found in the dataset.")
```

```python
# Analyze HangUpTime patterns and identify trends
if 'HangUpTime' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df['HangUpTime'], kde=True, bins=30)
    plt.title('Distribution of HangUpTime')
    plt.xlabel('HangUpTime (seconds)')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("Column 'HangUpTime' not found in the dataset.")
```

```python
# Determine the order status distribution for calls
if 'orderStatus' in df.columns:
    order_status_distribution = df['orderStatus'].value_counts()
    print("Order Status Distribution:")
    print(order_status_distribution)
else:
    print("Column 'orderStatus' not found in the dataset.")
```

```python
# Calculate the total refund amount and identify the refund status distribution
if 'refundStatus' in df.columns and 'refunds[0]' in df.columns:
    # Assuming 'refunds[0]' contains the refund amounts
    total_refund_amount = pd.to_numeric(df['refunds[0]'], errors='coerce').sum()
    refund_status_distribution = df['refundStatus'].value_counts()
    print(f"\nTotal Refund Amount: {total_refund_amount}")
    print("\nRefund Status Distribution:")
    print(refund_status_distribution)
else:
    print("Columns 'refundStatus' or 'refunds[0]' not found in the dataset.")
```

# Insights and Recommendations

```python
# Insights and Recommendations

# Calculate average connection time
if 'ConnectTime' in df.columns and 'DialTime' in df.columns:
    df['ConnectDelay'] = df['ConnectTime'] - df['DialTime']
    avg_connect_time = df['ConnectDelay'].mean()
    print(f"Average Connection Time: {avg_connect_time:.2f} seconds")
else:
    print("Connection time data is not available.")

# Analyze refund patterns
if 'refunds[0]' in df.columns and 'refundStatus' in df.columns:
    total_refunds = pd.to_numeric(df['refunds[0]'], errors='coerce').sum()
    refund_status_distribution = df['refundStatus'].value_counts()
    print(f"Total Refund Amount: {total_refunds}")
    print("Refund Status Distribution:")
    print(refund_status_distribution)
else:
    print("Refund data is not available.")

# Correlation between TalkTime and Charge
if 'timeDuration' in df.columns and 'amount' in df.columns:
    correlation = df[['timeDuration', 'amount']].corr().iloc[0, 1]
    print(f"Correlation between TalkTime and Charge: {correlation:.2f}")
else:
    print("TalkTime or Charge data is not available.")

# Analyze order status
if 'orderStatus' in df.columns:
    order_status_distribution = df['orderStatus'].value_counts(normalize=True) * 100
    print("Order Status Distribution (Percentage):")
    print(order_status_distribution)
else:
    print("Order status data is not available.")

# Analyze HangUpTime patterns
if 'HangUpTime' in df.columns:
    avg_hangup_time = df['HangUpTime'].mean()
    print(f"Average HangUpTime: {avg_hangup_time:.2f} seconds")
else:
    print("HangUpTime data is not available.")

# Recommendations summary
print("\nRecommendations:")
print("- Optimize connection routing to reduce delays.")
print("- Investigate reasons for refunds and address service quality gaps.")
print("- Leverage activities with higher TalkTime for revenue growth.")
print("- Improve processes for incomplete or canceled orders.")
print("- Analyze HangUpTime patterns to enhance reliability and satisfaction.")
```

```python

# Convert 'createdAt' to datetime if available and create line chart
if 'createdAt' in df.columns and 'amount' in df.columns:
    df['createdAt'] = pd.to_datetime(df['createdAt'], errors='coerce')
    df_sorted = df.sort_values(by='createdAt')
    call_charges_trend = df_sorted.groupby('createdAt')['amount'].sum()
    
    plt.figure(figsize=(12, 6))
    call_charges_trend.plot()
    plt.title('Trend of Call Charges Over Time')
    plt.xlabel('Date')
    plt.ylabel('Total Charges')
    plt.grid()
    plt.show()
else:
    print("Columns 'createdAt' or 'amount' not found in the dataset.")
```
